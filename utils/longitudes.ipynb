```python
import matplotlib.pyplot as plt
from os import listdir
import numpy as np
```

```python
def histograma(data, title, rows, cols, idx):
    plt.subplot(rows, cols, idx)
    plt.hist(data, bins='auto')
    plt.xlim(0,600)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title(title)

def read_lengths(file):
    lengths = []
    with open(file, 'r') as f:
        for line in f:
            lengths.append(len(line))
    return lengths
```

```python
enfr = read_lengths('../europarl-inmt/fr-en/test.en')
fren = read_lengths('../europarl-inmt/fr-en/test.fr')
ende = read_lengths('../europarl-inmt/de-en/test.en')
deen = read_lengths('../europarl-inmt/de-en/test.de')
enes = read_lengths('../europarl-inmt/es-en/test.en')
esen = read_lengths('../europarl-inmt/es-en/test.es')
```

```python
means = [np.mean(enfr), np.mean(fren), np.mean(ende), np.mean(deen), np.mean(enes), np.mean(esen)]
stds = [np.std(enfr), np.std(fren), np.std(ende), np.std(deen), np.std(enes), np.std(esen)]
maxs = [np.max(enfr), np.max(fren), np.max(ende), np.max(deen), np.max(enes), np.max(esen)]

plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
plt.errorbar(np.arange(6),means, yerr=stds, fmt='o')
plt.xticks(np.arange(6), ['enfr', 'fren', 'ende', 'deen', 'enes', 'esen'])
plt.title('Medias')

plt.subplot(2,3,2)
plt.bar(np.arange(6),stds)
plt.xticks(np.arange(6), ['enfr', 'fren', 'ende', 'deen', 'enes', 'esen'])
plt.title('Desviaciones tipicas')

plt.subplot(2,3,3)
plt.plot(np.arange(6),maxs, 'o')
plt.xticks(np.arange(6), ['enfr', 'fren', 'ende', 'deen', 'enes', 'esen'])
plt.title('Maximos')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
histograma(enfr, 'EN-FR', 2, 3, 1)
histograma(fren, 'FR-EN', 2, 3, 4)
histograma(ende, 'EN-DE', 2, 3, 2)
histograma(deen, 'DE-EN', 2, 3, 5)
histograma(enes, 'EN-ES', 2, 3, 3)
histograma(esen, 'ES-EN', 2, 3, 6)
plt.show()
```

```python
threshold = 512
overflow = [np.count_nonzero(np.array(enfr) > threshold), np.count_nonzero(np.array(fren) > threshold), np.count_nonzero(np.array(ende) > threshold), np.count_nonzero(np.array(deen) > threshold), np.count_nonzero(np.array(enes) > threshold), np.count_nonzero(np.array(esen) > threshold)]
plt.figure()
plt.bar(np.arange(6), overflow)
plt.xticks(np.arange(6), ['enfr', 'fren', 'ende', 'deen', 'enes', 'esen'])
plt.title('Overflow')
plt.show()
```
